```python
import pandas as pd
import os 
```

```python
vine_df = pd.read_csv(r'vine.csv')
```

```python
vine_df.tail()
```

```python
# sorting df to make sure I have values greater than 20. 
vine_df.sort_values(by='total_votes', ascending=False)
```

```python
# Retrieve all rows where the total_votes count is equal >=20 

total_votes = vine_df.loc[vine_df['total_votes'] >= 20]
total_votes.head()
```

```python
# Retrieve all the rows where the number of helpful_votes divided by total_votes is >= 50% 

df= pd.DataFrame(total_votes)
df["result"] = df["helpful_votes"]/df["total_votes"] * 100
df
```

```python
helpful_votes = df.loc[df['result'] >= 50]
helpful_votes
```

# PAID REVIEWS 


```python
# Retrieves all the rows where a review was written as part of the Vine Program, vine == 'Y'.
vine_programy = helpful_votes.loc[helpful_votes['vine'] == 'Y']
vine_programy
```

```python
total_paid_reviews = vine_programy.count()
total_paid_reviews
```

```python
five_star_paid = vine_programy.loc[vine_programy['star_rating'] == 5]
five_star_paid_count = five_star_paid.count()
```

```python
percentage_of_5_star_paid = five_star_paid_count/total_paid_reviews * 100
percentage_of_5_star_paid
```

# UNPAID REVIEWS 

```python
# Retrieves all the rows where a review was not written as part of the Vine Program, vine == 'N'.
vine_programn = helpful_votes.loc[helpful_votes['vine'] == 'N']
vine_programn
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for 
# the two types of review
total_unpaid_reviews = vine_programn.count()
total_unpaid_reviews
```
